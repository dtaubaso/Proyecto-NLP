## Modelo de clasificación

Vamos a generar un modelo que pueda predecir la puntuación de un comentario. Para ello, vamos a normalizar los textos, extraer solo las palabras más importantes y luego utilizar un algoritmo de clasificación sobre el nuevo dataset generado.

```python
import gc #garbage collector
gc.collect()
```

```python
#Importamos Spacy para procesar el dataset de palabras
import spacy
nlp = spacy.load("es_core_news_md")
```

```python
#Generamos los nuevos datasets
df_train["review"] = df_train["review_title"] +" "+ df_train["review_body"]
df_test["review"] = df_test["review_title"] +" "+ df_test["review_body"]
df_dev["review"] = df_dev["review_title"] +" "+ df_dev["review_body"]
```

```python
df_train.head()
```

```python
#Vamos a limpiar los datasets y dejar solo los features que nos interesan para este análisis
train = df_train[["stars","review"]]
test = df_test[["stars","review"]]
dev = df_dev[["stars","review"]]
```

Vamos a crear una función para normalizar cada uno de los datasets, para ello utilizaremos **Spacy** para tokenizar y lemmatizar.

```python
#El próximo paso es crear una función para normalizar cada uno de los datasets
def normalize(columna):
    normalizada = []
    for i in range(columna.shape[0]):
        review = columna[i] #seleccionar la review
        review = nlp(review)
        #tokenizar y lemmatizar con spacy, quitando signos de puntuación y números
        review = [tok.lemma_.lower() for tok in review if tok.pos_ != 'PUNCT' and tok.pos_ != 'NUM'] 
        review = [word for word in review if word not in stop_words] # Filtrar por stopwords
        review=" ".join(review) #unimos nuevamente la review
        normalizada.append(review) #agregar el resultado a la lista
    return normalizada
```

```python
#Ahora normalizamos el dataset de train
train_norm = normalize(train.review)
```

```python
#Vamos a generar un dataset para poder exportarlo por el tiempo que lleva normalizar
df_train_norm = pd.DataFrame({'reviews': train_norm,'stars': train.stars})
df_train_norm.head()
```

```python
#Exportamos a CSV el dataset
path = "C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\df_train_norm.csv"
df_train_norm.to_csv(path)
```

```python
#Normalizamos y exportamos el set de test
test_norm = normalize(test.review)
df_test_norm = pd.DataFrame({'reviews': test_norm,'stars': test.stars})
path = "C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\df_test_norm.csv"
df_test_norm.to_csv(path)
```

```python
#Normalizamos y exportamos el set de dev
dev_norm = normalize(dev.review)
df_dev_norm = pd.DataFrame({'reviews': dev_norm,'stars': dev.stars})
path = "C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\df_dev_norm.csv"
df_dev_norm.to_csv(path)
```

```python
df_train_norm = pd.read_csv("C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\df_train_norm.csv")
df_test_norm = pd.read_csv("C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\df_test_norm.csv")
df_dev_norm = pd.read_csv("C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\df_dev_norm.csv")
```

### Vectorizar

Hasta ahora, contamos con una **"bolsa de palabras"** ("bag of words"), pero debemos represntarlas como un vector numérico para poder modelar un algoritmo de clasificación. Para este paso utilizaremos la función CountVectorizer de Scikit-Learn, que permite representar cada vector con un 1 en las palabras que contiene y un 0 en las que no. Para simplificar la tarea y descartar errores, utilizaremos únicamente las palabras que más veces aparecen en el texto.

```python
#Importamos el algoritmo
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Usaremos solo las 2500 palabras con más frecuencia en todo el corpus para generar los vectores
#Esto se hace para eliminar palabras que aparecen poco y descartar errores

max_features=2500

#Es decir que cada instancia tendrá 2500 features

#Instanciamos
cv=CountVectorizer(max_features=max_features)
```

```python
#Entrenamos y transformamos el set de train
train_vec = cv.fit_transform(df_train_norm.reviews)
```

```python
#Vamos a chequear cuáles son las palabras que fueron vectorizadas
#Para ello, reconvertimos el vector a palabras nuevamente
all_words = cv.get_feature_names()

#Luego, visualizamos 50 palabras del set de manera aleatoria
import random
print("50 palabras al azar: ",random.sample(all_words, 50))
```

```python
#Vamos a vectorizar también el set de test, que será nuestro x_test
#Este set lo vamos a transformar, sin entrenar, para que tome los parámetros del set de train
test_vec = cv.transform(df_test_norm.reviews)
```

```python
#Ahora, separamos los datos en entrenamiento y test.
x_train = train_vec.toarray()
x_test = test_vec.toarray()
y_train = df_train_norm.stars
y_test = df_test_norm.stars
```

### Modelar

Ya que tenemos los sets de entrenamiento y test, podemos comenzar a probar distintos modelos de Machine Learning para ver cuál funciona mejor en este problema de clasificación.

#### Naive Bayes
Primero vamos a entrenar un modelo de Naive Bayes para usar como **Benchmark**. Para medir el desempeño utilizaremos el **Score** propio de cada algoritmo, que calcula la precisión media, comparando las etiquetas predichas con las verdaderas. El puntaje varía de 0 a 1, siendo 1 el número que se obtiene si todas las etiquetas coinciden.

También usaremos la métrica **F1-Score**, que puede interpretarse como una media ponderada entre la **precisión** (proporción de todas las predicciones positivas que son correctas - exactitud) y la **sensibilidad** (proporción de observaciones positivas reales que se predijeron correctamente - exhaustividad), donde la puntuación F1 alcanza su mejor valor en 1 y la peor puntuación en 0. 

Por último, graficaremos una **Matriz de Confusión**, en la que cada columna de la matriz representa el número de predicciones de cada clase, mientras que cada fila representa a las instancias en la clase real. Esta gráfica facilita ver si el sistema está confundiendo dos clases.

```python
#Importamos las métricas
from sklearn.metrics import plot_confusion_matrix,f1_score
```

```python
#Importamos el algoritmo de Naive Bayes
from sklearn.naive_bayes import GaussianNB
```

```python
#Instanciamos y entrenamos
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
#Veamos la precisión de este algoritmo
print("Precisión: ", nb.score(x_test,y_test))
```

La precisión del algoritmo es muy baja, menos de 0.5. Veremos qué está sucediendo con el F1 Score

```python
#Vamos a ver qué ocurre al predecir
y_pred = nb.predict(x_test)
```

```python
#Veamos el score F1
print("F1-Score: ",f1_score(y_test, y_pred, average=None))
```

Al analizar el F1-Score, podemos ver que la precisión **baja mucho en las etiquetas intermedias**, mientras que en los extremos los números son sensiblemente más altos, aunque no tan buenos.

```python
#Graficamos una matriz de confusión
fig, ax = plt.subplots(figsize=(10, 10))
cm = plot_confusion_matrix(nb, x_test, y_test, ax=ax)
fig.suptitle("Matriz de confusión")
plt.grid(False)
plt.show()
```

La **matriz de confusión** permite ver lo dicho anteriormente: el modelo tiene problemas para predecir las etiquetas intermedias de la clasificación. Veremos si con los siguientes algoritmos esto mejora.

#### Random Forest
Veamos cómo funciona un modelo de random forest para este problema de clasificación.

```python
#Importamos el algoritmo
from sklearn.ensemble import RandomForestClassifier
```

```python
#Instanciamos y entrenamos
rf = RandomForestClassifier(random_state=0)
rf.fit(x_train,y_train)
```

```python
#Veamos la precisión de este modelo
print("Precisión: ", rf.score(x_test,y_test))
```

```python
#Vamos a ver qué ocurre con el F1-Score
y_pred = rf.predict(x_test) #Predecimos
#Imprimimos el score F1
print("F1-Score: ",f1_score(y_test, y_pred, average=None))
```

De manera similar al modelo anterior, podemos ver que en las etiquetas de los extremos (1 y 5) la predicción es sutilmente mejor que en las intermedias.

Si bien este modelo mejora levemente respecto al anterior, podemos ver también que tiene dificultad para predecir las **etiquetas intermedias** ( que en este dataset corresponde a calificaciones intermedias entre positivas y negativas).

#### Importancia de los features

Una característica muy interesante de Random Forest es poder preguntarle por la **importancia de los features** que usó para clasificar. Veamos en este caso cuales son las palabras que mayormente determinan la clasificación.

```python
# Le preguntamos la importancia de cada feature (cada palabra)
importances = rf.feature_importances_
# Tomamos la lista de palabras
columns = all_words

# Ordenamos por importancia  y tomamos las 20 primeras
indices = np.argsort(importances)[::-1] #Ordena los valores según índice en orden decreciente
indices = indices[:20] #Toma los primeros 20 valores
selected_columns = [columns[i] for i in indices] #Le asigna la palabra a cada valor
selected_importances = importances[indices] #Ordena los valores de y

# Por último, graficamos
plt.figure(figsize = (15,8))
sns.barplot(selected_columns, selected_importances)
plt.show()
```

Una pista que podemos concluir de esta gráfica es que tal vez tomar una sola palabra para clasificar no sea suficiente, porque la palabra "no" es sumamente ambigua. Tal vez convenga **aumentar el número de n-gramas** al realizar la vectorización.

# Problema binario

Todo apunta a pensar que una buena manera de encarar este problema de clasificación sería verlo como un **problema binario**, es decir, dividir el dataset en dos grandes grupos de reviews **"positivas"** y **"negativas"**. A juzgar por los resultados obtenidos, en los que las mejores predicciones se obtuvieron en los dos extremos de las calificaciones, habría que ver si binarizar el dataset no supone una mejora en la precisión.

### Dividir el dataset

Vamos a generar dos listas, de reviews "positivas" y "negativas". Utilizaremos los sets normalizados para no tener que volver a pasar por todo el proceso. Para ello, tomamos las reviews calificadas con "1" y "2" para el set de "negativas" y las calificadas con "4" y "5" para el set de "positivas". Vamos a descartar en esta etapa aquellas reviews con calificación de "3", por ser demasiado ambiguas para este análisis. Asimismo, vamos a calificar con "0" a las reviews negativas y con "1" a las positivas.

```python
#Cargamos nuevamente los sets
df_train_norm = pd.read_csv("C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\df_train_norm.csv")
df_test_norm = pd.read_csv("C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\df_test_norm.csv")
df_dev_norm = pd.read_csv("C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\df_dev_norm.csv")
```

```python
#separamos los sets en "positivos" y "negativos"
train_neg = df_train_norm[(df_train_norm.stars == 1)|(df_train_norm.stars == 2)]
train_pos = df_train_norm[(df_train_norm.stars == 4)|(df_train_norm.stars == 5)]
test_neg = df_test_norm[(df_test_norm.stars == 1)|(df_test_norm.stars == 2)]
test_pos = df_test_norm[(df_test_norm.stars == 4)|(df_test_norm.stars == 5)]
dev_neg = df_dev_norm[(df_dev_norm.stars == 1)|(df_dev_norm.stars == 2)]
dev_pos = df_dev_norm[(df_dev_norm.stars == 4)|(df_dev_norm.stars == 5)]
```

```python
#Vamos a agregar una columna de "0" o de "1" a cada dataset, según sea el caso
train_neg["sentiment"] = np.zeros((len(train_neg,)),dtype=int)
test_neg["sentiment"] = np.zeros((len(test_neg,)),dtype=int)
dev_neg["sentiment"] = np.zeros((len(dev_neg,)),dtype=int)
train_pos["sentiment"] = np.ones((len(train_pos,)),dtype=int)
test_pos["sentiment"] = np.ones((len(test_pos,)),dtype=int)
dev_pos["sentiment"] = np.ones((len(dev_pos,)),dtype=int)
```

```python
#Ahora vamos a unir cada dataset en un nuevo set
train_bi = pd.concat([train_pos, train_neg])
test_bi = pd.concat([test_pos, test_neg])
dev_bi = pd.concat([dev_pos,dev_neg])
```

```python
#Vamos a limpiar los datasets
train_bi.drop(["Unnamed: 0", "stars"], axis = 1, inplace=True)
test_bi.drop(["Unnamed: 0", "stars"], axis = 1, inplace=True)
dev_bi.drop(["Unnamed: 0", "stars"], axis = 1, inplace=True)
```

```python
#Reseteamos el index
train_bi.reset_index(inplace=True, drop=True)
test_bi.reset_index(inplace=True, drop=True)
dev_bi.reset_index(inplace=True, drop=True)
```

```python
#Vamos a exportar los datasets como csv
path_train = "C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\train_bi.csv"
path_test = "C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\test_bi.csv"
path_dev = "C:\\Users\\dtaub\\Documents\\Data_Science\\proyecto_3\\dev_bi.csv"
train_bi.to_csv(path_train)
test_bi.to_csv(path_test)
dev_bi.to_csv(path_dev)
```

## Vectorizar

Vamos a vectorizar estos nuevos datasets, con un rango de 3 ngramas que dio un mejor resultado.

```python
# Nuevamente usaremos solo las 2500 palabras con más frecuencia del corpus
max_features=2500

#Instanciamos
cv_bi=CountVectorizer(max_features=max_features, ngram_range=(1,3))

#Entrenamos y transformamos el set de train
train_vec_bi = cv_bi.fit_transform(train_bi.reviews)
```

```python
#Vamos a vectorizar también los sets de test y dev
test_vec_bi = cv_bi.transform(test_bi.reviews)
dev_vec_bi = cv_bi.transform(dev_bi.reviews)
```

## Modelar

Ahora sí, vamos a ver qué ocurre al entrenar un algoritmo de Machine Learning con esta nueva disposición.

```python
#Ahora, separamos los datos en entrenamiento, test y dev.
xtrain_bi = train_vec_bi.toarray()
xtest_bi = test_vec_bi.toarray()
xdev_bi = test_dev_bi.toarray()
ytrain_bi = train_bi.sentiment
ytest_bi = test_bi.sentiment
ydev_bi = dev_bi.sentiment
```

#### SVC

```python
#Vamos a ver cómo responde el algoritmo de SVC

svc_bi = LinearSVC(C=1,random_state=1,dual=False)
svc_bi.fit(xtrain_bi,ytrain_bi)

print("Precisión: ", svc_bi.score(xtest_bi,ytest_bi))
```

Como vemos, la precisión **aumentó drásticamente** con este nuevo enfoque. Vamos a ver cómo responden las demás métricas.

```python
#Analizamos el F1-Score
ypred_test = svc_bi.predict(xtest_bi) 
ypred_dev = svc_bi.predict(xdev_bi)

print("F1-Score test: ",f1_score(ytest_bi, ypred_test))
print("F1-Score dev: ",f1_score(ydev_bi, ypred_dev))
```

Nuevamente, los puntajes resultan ser **increíblemente superiores** a los ejemplos vistos anteriormente. Vamos a ver qué ocurre con la matriz de confusión.

```python
#Graficamos una matriz de confusión
fig, ax = plt.subplots(figsize=(10, 10))
cm = plot_confusion_matrix(svc_bi, xtest_bi, ytest_bi, ax=ax)
fig.suptitle("Matriz de confusión")
plt.grid(False)
plt.show()
```

La matriz de confusión nos muestra cómo menos del 10% de cada etiqueta fue identificado de forma errónea por este modelo, lo cual da un excelente resultado, aunque hace sospechar que puede haber **overfitting**, por lo que vamos a probar una técnica de **validación cruzada** y **Kfold** para mezclar los datos y ver si difieren los resultados.

```python
#Importamos las librerías
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

#Creamos un fold con datos mezclados
kf = KFold(shuffle=True, n_splits=10)

#Probamos la puntuación
scores = cross_val_score(svc_bi, xtrain_bi, ytrain_bi, cv=kf)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
#Hacemos lo mismo en los sets de test y dev

scores = cross_val_score(svc_bi, xtest_bi, ytest_bi, cv=kf)
print("Accuracy test: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

scores = cross_val_score(svc_bi, xdev_bi, ydev_bi, cv=kf)
print("Accuracy dev: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

Como vemos, los resultados están en el orden de 0.9 de precisión, un número dramáticamente superior al que se venía obteniendo al intentar predecir valores de 1 a 5. 

Como última comprobación, vamos a ver qué ocurre con el algoritmo elegido como **benchmark** al principio de este notebook.

#### Naive Bayes


```python
#Instanciamos y entrenamos nuevamente el algoritmo

nb_bi = GaussianNB()
nb_bi.fit(xtrain_bi, ytrain_bi)

#Veamos la precisión de este algoritmo
print("Precisión: ", nb_bi.score(xtest_bi,ytest_bi))
```

Tal como ocurrió con el modelo anterior, **la precisión aumentó radicalmente**, pasando de 0.34 a 0.85, un número más que adecuado para pensar en un modelo que realmente sirva para predecir el tono de una predicción.

```python
#Analizamos el F1-Score
ypred_test = nb_bi.predict(xtest_bi) 
ypred_dev = nb_bi.predict(xdev_bi)

print("F1-Score test: ",f1_score(ytest_bi, ypred_test))
print("F1-Score dev: ",f1_score(ydev_bi, ypred_dev))
```

Los resultados de F1-Score están en línea con los obtenidos con la métrica de precisión, también con una marcada mejoría en este nuevo dataset.

```python
#Por último, graficamos una matriz de confusión
fig, ax = plt.subplots(figsize=(10, 10))
cm = plot_confusion_matrix(nb_bi, xtest_bi, ytest_bi, ax=ax)
fig.suptitle("Matriz de confusión")
plt.grid(False)
plt.show()
```

La matriz de confusión permite ver que el modelo ahora acierta la clasificación con muchísima mayor exactitud.
